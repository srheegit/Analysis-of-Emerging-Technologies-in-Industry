## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Sample of data

QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
import matplotlib.pyplot as plt

# Select the converted compensations of software engineers
QUERY = """
SELECT ConvertedComp
FROM master
"""

# Run the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)
# df.head()

# Plot the result as a histogram
plt.hist(df['ConvertedComp'].dropna(), bins='auto', color = 'blue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""

# Run the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Plot the result as a boxplot
plt.boxplot(df['Age'].dropna(), vert=True)
plt.title('Box plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Run the query and store the result in a pandas dataframe
df = pd.read_sql_query(QUERY, conn)

# Plot a scatterplot
plt.scatter(df['Age'], df['WorkWeekHrs'])
plt.title('Scatterplot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Run the query and store the result in a pandas dataframe
df = pd.read_sql_query(QUERY, conn)

# Remove Na values
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# plot the bubble plot
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], alpha=0.5, s=df['Age']*10)
plt.title('Bubble plot of WorkWeekHrs vs. CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Query to get the 'DatabaseNextYear' column data
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# Run the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each database
database_counts = df['DatabaseDesireNextYear'].value_counts()

# Get the top 5 databases
top_databases = database_counts.head(5)

# Create a pie chart
plt.figure(figsize=(10,6))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Top 5 databases respondents wish to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Query to get the 'WorkWeekHrs', 'CodeRevHrs' and 'Age' column data
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Run the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Calculate the median of 'WorkWeekHrs' and 'CodeRevHrs' for the age group 30 to 35
median_work_week_hrs = df['WorkWeekHrs'].median()
median_code_rev_hrs = df['CodeRevHrs'].median()

# Create a list of the median values
medians = [median_work_week_hrs, median_code_rev_hrs]

# Create a stacked bar chart
plt.bar(['WorkWeekHrs', 'CodeRevHrs'], medians, color=['blue', 'orange'])
plt.ylabel('Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for age group 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Query to get the 'ConvertedComp' and 'Age' column data for ages 45 to 60
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Run the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Group by 'Age' and calculate the median 'ConvertedComp'
median_comp_by_age = df.groupby('Age')['ConvertedComp'].median()

# Plot the median 'ConvertedComp' for all ages from 45 to 60
plt.plot(median_comp_by_age.index, median_comp_by_age.values)
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Query to get the 'MainBranch' column data
QUERY = """
SELECT MainBranch
FROM master
"""

# Run the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each category in 'MainBranch'
mainbranch_counts = df['MainBranch'].value_counts()

# Create a horizontal bar chart
mainbranch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Horizontal bar chart of MainBranch')
plt.show()
```
