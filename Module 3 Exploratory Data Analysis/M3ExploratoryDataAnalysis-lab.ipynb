## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python
import matplotlib.pyplot as plt
import numpy as np

data = df['ConvertedComp'].fillna(0)
# Create histogram
counts, bin_edges = np.histogram(data, bins='auto', density=True)

# Calculate bin centers
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Plot the density curve
plt.plot(bin_centers, counts, linewidth=3)

# Plot formatting
plt.title('Density Plot for ConvertedComp')
plt.xlabel('ConvertedComp (Annual USD salaries)')
plt.ylabel('Density')

plt.show()
```

Plot the histogram for the column `ConvertedComp`.


```python
plt.hist(data, bins = 'auto', edgecolor = 'black')

plt.title('Histogram for Converted Comp')
plt.xlabel('Converted Comp')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
import matplotlib.pyplot as plt

data = df['ConvertedComp']

# Create boxplot
plt.boxplot(data.dropna(), vert=False)

# Plot formatting
plt.title('Box Plot for ConvertedComp')
plt.xlabel('ConvertedComp (Annual USD salaries)')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
data = df['ConvertedComp'].dropna()

# Calculate Q1, Q3, and IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

Find out the upper and lower bounds.


```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f'IQR: {IQR}')
print(f'Lower Bound: {lower_bound}')
print(f'Upper Bound: {upper_bound}')
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
num_outliers = len(data[data < lower_bound]) + len(data[data > upper_bound])
print("There are {} outliers in our data".format(num_outliers))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
data = data[(data >= lower_bound) & (data <= upper_bound)]
```

```python
# Median ConvertedComp after removing outliers
print("The median converted compensation after removing outliers is", data.median())
```

```python
# Mean ConvertedComp after removing outliers
print("The mean converted compensation after removing outliers is", data.mean())
```
